```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import pandas
import numpy 
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from matplotlib.pyplot import subplot
```

```python
dataset = pandas.read_csv('frb_01_overall_new2.csv')
X = dataset.iloc[:,17:18].values  # % Borrowers without a Pell Grant
y = dataset.iloc[:, 20].values  # Mean Balance

Xy = numpy.array(list(zip(X,y)))
kmeans = KMeans(n_clusters = 5)
kmeans.fit(Xy)
centers = kmeans.cluster_centers_
labels = kmeans.labels_
colors = ['r.','g.','b.','c.','y.','m.']
#matplotlib.pyplot.scatter(X,y)
print('Displaying: % Borrowers without a Pell Grant vs. Mean Balance')

for i in range(len(Xy)):
    plt.plot(Xy[i][0],Xy[i][1],
                           colors[labels[i]],
                           markersize=10)
    #print(i)
    
plt.title('% Borrowers without a Pell Grant vs. Mean Balance')
plt.xlabel('% Borrowers without a Pell Grant')
plt.ylabel('Mean Balance')
#plt.scatter(centers[:, 0], centers[:, 1], c='black', s=100,alpha=0.85)
plt.show()
```

```python
for i in range(0,75):
    print('School: '+ dataset.iloc[i:i+1,1].values + ' Label: '+ str(labels[i]) +' Risk: ' +dataset.iloc[i:i+1,3].values)
```

```python
print('Displaying: Median Dependent Parent AGI vs. Mean Balance')

X = dataset.iloc[:,14].values  # Median Dependent Parent AGI
Xy = numpy.array(list(zip(X,y)))
kmeans = KMeans(n_clusters = 5)
kmeans.fit(Xy)
centers = kmeans.cluster_centers_
labels = kmeans.labels_
colors = ['r.','g.','b.','c.','y.','m.']
#plt.scatter(centers[:, 0], centers[:, 1],zorder=-1, c='black', s=200,alpha=0.85)
for i in range(len(Xy)):
    plt.plot(Xy[i][0],Xy[i][1],
                           colors[labels[i]],
                           markersize=10)
    
plt.title('Displaying: Median Dependent Parent AGI vs. Mean Balance')
plt.xlabel('Displaying: Median Dependent Parent AGI')
plt.ylabel('Mean Balance')
plt.show()
```
