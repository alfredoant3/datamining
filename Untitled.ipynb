```python
import matplotlib.pyplot
from matplotlib import style
#style.use('ggplot')
import pandas
import numpy 
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from scipy.stats import mode
```

```python
dataset = pandas.read_csv('frb_01_overall_new2.csv')
X = dataset.iloc[:,17].values  # % Borrowers without a Pell Grant
y = dataset.iloc[:, 20].values  # Mean Balance
```

```python
Xy = numpy.array(list(zip(X,y)))
Xy
```

```python
k_means = KMeans(n_clusters = 6)
k_means.fit(Xy)
ykmeans = k_means.predict(Xy)
labels = k_means.labels_
center = k_means.cluster_centers_
colors = ['red','green','blue','cyan','yellow','magenta']


#for i in range(len(Xy)):
    #matplotlib.pyplot.plot(X[i],y[i],markersize = 10)
    #print('iter: '+str(i))
matplotlib.pyplot.scatter(X,y)
matplotlib.pyplot.scatter(center[:,0], center[:,1], c='black',s=100,alpha=0.85)
matplotlib.pyplot.show()   
```
